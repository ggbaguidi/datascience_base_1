```python
commande = pd.read_csv('commande.csv')
produits = pd.read_csv('produits.csv')
data = pd.merge(commande, produits, on='id', how='inner')
data['CA'] = data['nombre'] * data['prix']
data_grouped = data.groupby('id').sum()
```
